## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(x=df["ConvertedComp"])
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
df["ConvertedComp"].describe()
```

Find out the upper and lower bounds.


```python
# your code goes here
print(df["ConvertedComp"].max(), df["ConvertedComp"].min())
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
Q1 = df["ConvertedComp"].quantile(0.25)
Q3 = df["ConvertedComp"].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
outlier_count = 0
for row in df["ConvertedComp"]:
    if row > upper_bound or row < lower_bound:
        outlier_count += 1
print(outlier_count)
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
# your code goes here
Q1 = df["ConvertedComp"].quantile(0.25)
Q3 = df["ConvertedComp"].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
filtered_df = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
filtered_df
```

```python
filtered_df["ConvertedComp"].median()
```

```python
filtered_df["ConvertedComp"].mean()
```
